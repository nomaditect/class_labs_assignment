## Import csv files

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
# check info
file1.info()
```

```python
# change file1 column names to lowercase.

cols = []
for i in range(len(file1.columns)):
    cols.append(file1.columns[i].lower())
    
file1.columns = cols
file1.info()
```

```python
#replace st with state in file1.

file1.rename(columns={'st': 'state'}, inplace=True)
file1.info()
```

```python
# change file2 column names to lowercase.

cols2 = []
for i in range(len(file2.columns)):
    cols2.append(file2.columns[i].lower())

file2.columns = cols2
file2.info()
```

```python
#replace st with state file2.
file2.rename(columns={'st': 'state'}, inplace=True)
file2.info()
```

```python
# change file3 column names to lowercase.

cols3 = []
for i in range(len(file3.columns)):
    cols3.append(file3.columns[i].lower())

file3.columns = cols3
file3.info()
```

## Create new data frame

```python
columns = file1.columns

data = pd.DataFrame(columns=columns)
data
```

## Concatenate files together

```python
data = pd.concat([data,file1,file2,file3],axis=0)
```

```python
#change 'number of open complaints' to float.

#data[['number of open complaints']].astype(int)
#didnt work with nul values.

data['number of open complaints']=pd.to_numeric(data['number of open complaints'],errors='coerce')
data.info()
```

## reindex the data

```python
data.reset_index(drop=True,inplace=True)
data.info()
```

```python
data
```

## Delete column : education + number of open complaints

```python
datanew = data.drop(['education'],axis=1)
datanew = data.drop(['number of open complaints'],axis=1)
datanew
```

```python
datanew.dtypes
```

```python
# Correct the values in the column customer lifetime value. 
# They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.

datanew['customer lifetime value']*=100
datanew['customer lifetime value']=pd.to_numeric(datanew['customer lifetime value'],errors='coerce')
```

```python
#delete duplicates
datanew=data.drop_duplicates()
datanew.shape
```
